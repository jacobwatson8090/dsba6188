```python
def read_sentences_from_file(file_path):
    sentences = []

    with open(file_path, 'r') as file:
        # Read the content of the file
        text = file.read()

    # Split the text into sentences based on periods
    sentences = text.split('.')

    # Remove empty strings resulting from split() method
    sentences = [sentence.strip() for sentence in sentences if sentence.strip()]

    return sentences
```

# Consumer Goods/Retail Topic Model & Topic Map

```python
cgr_path = "/content/ConsumerGoods-RetailConsolidated.txt"

sentences = read_sentences_from_file(cgr_path)
```

```python
from bertopic import BERTopic

cgr_topic_model = BERTopic(min_topic_size=20)
topics, probs = cgr_topic_model.fit_transform(sentences)
```

```python
# setting nr_topics to auto reduced number of topics from 80 to 5

cgr_topic_model.get_topic_info()
```

```python
cgr_topic_model.visualize_topics()
```
